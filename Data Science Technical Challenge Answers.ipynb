#### 1. Think about what could be going wrong with our calculation. Think about a better way to evaluate this data. 

The AOV of 3145.13 dollars was most likely gotten by just taking the mean of the order_amount column. This is problematic because of outliers like 704000 dollars. This user could be buying for wholesale purposes and would not reflect the buying behaviour of regular sneaker purchases. The Standard Deviation of the Order Amount is 41282.54 dollars, this shows that most of the values are very far from the average. Thus while the average was calculated properly, it does not give an accurate depiction of the average buyer. 

#### 2. What metric would you report for this dataset?

A better metric would be the Unit Order Amount. The largest Order Amount came from an Order of 2000 items, this evaluates to a unit order amount of 352 dollars. The Unit Order Amount scales the amount from each buyer, thus providing a more accurate depiction of the average buyer. It also has a much smaller standard deviation of 2441.96 dollars.

#### 3. What product was ordered the most by customers in Germany?

SELECT PID, PName, MAX(NOrders) 
FROM (SELECT Products.ProductID AS PID, Customers.CustomerID AS CID, Products.ProductName AS PName, COUNT(*) AS NOrders
FROM [Customers]
JOIN Orders ON Customers.CustomerID = Orders.CustomerID
JOIN OrderDetails ON Orders.OrderID = OrderDetails.OrderID
JOIN Products ON OrderDetails.ProductID = Products.ProductID
WHERE Country == 'Germany'
GROUP BY Products.ProductID)

Ans = Gorgonzola Telino
